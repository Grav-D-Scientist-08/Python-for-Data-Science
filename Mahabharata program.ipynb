```python
import random

def introduction():
    print("Welcome to the Mahabharata Manager Simulation!")
    print("You are playing the role of Lord Krishna, the manager of the Pandavas.")
    print("Your decisions will shape the outcome of the Mahabharata war.")

def choose_action():
    print("\nChoose an action:")
    print("1. Give advice to Arjuna.")
    print("2. Counsel Yudhishthira on strategy.")
    print("3. Help Bhima with his strength.")
    print("4. Offer guidance to Nakula and Sahadeva.")
    print("5. Abhimanu and the chakravyuh  ")
    print("6. Exit")
    return input("Enter the number of your choice: ")

def give_advice_to_arjuna():
    advice = ["Focus on your duty: Arjuna.\nKeep your mind stead: Arjuna." "\nRemember your training: Arjuna."]
    print(random.choice(advice))

def counsel_yudhishthira():
    strategy = ["Stay true to dharma: Yudhishthira\nPlan your moves carefully: Yudhishthira\nConsider peace talks:Yudhishthira."]
    print(random.choice(strategy))

def help_bhima_with_strength():
    strength_help = ["Use your immense power wisely:Bhima\nChannel your anger into strength: 4Bhima\nStay determined: Bhima."]
    print(random.choice(strength_help))

def offer_guidance_to_nakula_and_sahadeva():
    guidance = ["Support your brothers:Nakula and Sahadeva\nUse your skills to contribute:Nakula and Sahadeva.\nStay united and loyal: Nakula and Sahadeva."]
    print(random.choice(guidance))
    
def abhimanyu_and_the_chakravyuh():
    chakravyuh = ["The narration he heard from inside his mother's womb : Abhimanyu\n The battle at Kurukshetra:Abhimanyu "]
    print(random.choice(chakravyuh))
def main():
    introduction()
    while True:
        choice = choose_action()
        if choice == "1":
            give_advice_to_arjuna()
        elif choice == "2":
            counsel_yudhishthira()
        elif choice == "3":
            help_bhima_with_strength()
        elif choice == "4":
            offer_guidance_to_nakula_and_sahadeva()
        elif choice == "5":
            abhimanyu_and_the_chakravyuh()
        elif choice == "6":
            print("Thank you for playing the Mahabharata Manager Simulation!")
            break
        else:
            print("Invalid choice. Please select a valid option.")
```

```python
main()
```

```python
import turtle
from turtle import*
 
#screen for output
screen = turtle.Screen()
screen.title("@python.coder_ ")
 
# Defining a turtle Instance
t = turtle.Turtle()
speed(0)

 
# initially penup()
t.penup()
t.goto(-400, 250)
t.pendown()
 
# Orange Rectangle 
#white rectangle
t.color("orange")
t.begin_fill()
t.forward(800)
t.right(90)
t.forward(167)
t.right(90)
t.forward(800)
t.end_fill()
t.left(90)
t.forward(167)
 
# Green Rectangle
t.color("green")
t.begin_fill()
t.forward(167)
t.left(90)
t.forward(800)
t.left(90)
t.forward(167)
t.end_fill()
 
# Big Blue Circle
t.penup()
t.goto(70, 0)
t.pendown()
t.color("navy")
t.begin_fill()
t.circle(70)
t.end_fill()
 
# Big White Circle
t.penup()
t.goto(60, 0)
t.pendown()
t.color("white")
t.begin_fill()
t.circle(60)
t.end_fill()
 
# Mini Blue Circles
t.penup()
t.goto(-57, -8)
t.pendown()
t.color("navy")
for i in range(24):
    t.begin_fill()
    t.circle(3)
    t.end_fill()
    t.penup()
    t.forward(15)
    t.right(15)
    t.pendown()
     
# Small Blue Circle
t.penup()
t.goto(20, 0)
t.pendown()
t.begin_fill()
t.circle(20)
t.end_fill()
# Spokes
t.penup()
t.goto(0, 0)
t.pendown()
t.pensize(2)
for i in range(24):
    t.forward(60)
    t.backward(60)
    t.left(15)
     
#to hold the 
#output window
turtle.done()
```
