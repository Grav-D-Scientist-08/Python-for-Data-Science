# Bar Chart

We make bar chart when we have the catogorecal data.

```python
# Let's creat abar graph
fig,ax = plt.subplots(figsize=(5,5))
x = ["Shampoo","Oil","Soap"]
y = [1234,200,800]
ax.bar(x,y)
```

```python
help(ax.bar)
```

```python
#
fig,ax = plt.subplots(figsize=(5,5))
x = ["Shampoo","Oil","Soap"]
y = [1234,200,800]
ax.bar(x,y)
ax.set_xlabel("Product")
ax.set_ylabel("Units Sold")
ax.set_title("Units sold for product")
ax.grid(True)  # Shows grid
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lets change the colour also
fig,ax = plt.subplots(figsize=(5,5))
x = ["Shampoo","Oil","Soap"]
y = [1234,200,800]
color_bar = ["red","blue","green"]
ax.set_xlabel("Product") 

ax.set_ylabel("Units Sold")
ax.set_title("Units sold for product")
ax.grid(False)  # Shows grid
ax.bar(x,y,color = color_bar)
```

# Histogram

```python
fig,ax = plt.subplots(figsize=(10,5))
np.random.seed(12)
a = np.random.randint(100,size=80)   #Generates 80 random int between 0 to 100(exclusive)
ax.hist(a,bins=10)
ax.set_title("Random Generation")
ax.grid(True)
```

```python
fig,ax = plt.subplots(figsize=(10,5))
np.random.seed(12)
a = [20,30,25,12,55,40,50]  
ax.hist(a,bins=10)
ax.set_xlabel("Age")
ax.set_ylabel("freq")
ax.set_title("Random Generation")
```

```python
# Another way to do plotting
x = ["Shampoo","Oil","Soap"]
y = [1234,200,800]
color_bar = ["red","blue","green"]
plt.xlabel("Products")
plt.ylabel("Sale")
plt.title("Product vs Sale")
plt.bar(x,y,color = color_bar)
```

```python
# Pie chart 
company = ["Samsung","Redmi","OnePlus","Oppo"]
size = [300,20,15,35]
fig,ax = plt.subplots(figsize = (5,5))
ax.pie(size,labels = company, autopct = "%1.2f%%")
ax.set_title("Market Share")
```

```python
# Pie chart 
company = ["Samsung","Redmi","OnePlus","Oppo"]
size = [300,20,15,35]
fig,ax = plt.subplots(figsize = (5,5))
ax.pie(size,labels = company, autopct = "%25.3f%%") 
ax.set_title("Market Share")
```
