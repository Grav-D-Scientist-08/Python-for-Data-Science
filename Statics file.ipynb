# How to create a box and whisker plot? 
    
# step 1 :- Organize your data
    
   **Collect data** :- Insure you have all the data points you want to analyze.
    
   **Sort data** :- Arrange the data in a scending order.
    
# step 2 :- Find the five number summary 
   **Minimum** :- The smallest data point between Q1 and Q1 - 1.5 * IQR<br>
   **First Quartile (Q1)** :- The median of the lower half of the data set (Excluding the median if there's an odd number of data points).<br>
   **Median (Q2)** :- The middle value of the data set. If  there's an odd number data points.<br>
   **Third Quartile (Q3)** :- The median of the upper half of the dataset.<br>
   **Maximum** :- The largest data point between Q3 + 1.5 * IQR<br>
   <br>
# step 3 :- Identify Outliers 
   **Calulate the interquantile Range (IQR)**<br>
   **IQR** = Q3 - Q1<br>
   **Lower Boundary** = Q1 - 1.5 * IQR<br>
   **Upper Boundary** = Q3 + 1.5 * IQR<br>
   **Identify Outliers** :- Any data points outside these boundaries are considered outlires.<br>
# Step 4 :- Draw the box plot 
   **Draw a numbers line or scale** :- This should cover the range of your data.
   

# Draw the box:
 - The bottom of the box is at Q1.
 - The top of the box is at Q3.


# Add the whiskers :-
- Draw a line (whisker) from Q1 down to the minimum value which is up to the range og Q1 - 1.5 * IQR.
- Draw a line from Q3 up to the maximum value which is in the range of Q3 + 1.5 * IQR
# Plot outlires (if any) :- 
- Use dots or stars to mark any outlires beyond the whisker.
