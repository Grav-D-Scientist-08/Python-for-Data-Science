Feature scaling is a technique used in machine learning to normalize the range of features within a dataset.<br>
Feature scaling does not change the shape of the distribution of the data

### Why feature scaling is required?

- Smoother Convergence: Many machine learning algorithms, particularly those that use gradient descent to optimize<br>
  their parameters, take small steps in the direction that minimizes the error. The size of these steps is<br> often determined by the feature values. Having features on a similar scale will help the gradient descent converge more quickly<br> towards the minima.
- Distance-based algorithms: Distance-based algorithms like KNN, K-means, and SVM are most affected by the range<br> of features. This is because behind the scenes they are using distances between data points to determine their similarity and hence<br> perform the task at hand.

#### Algorithms requiring the use of feature scaling-
1. Linear Regression
2. Logistic Regression
3. KNN
4. Clustering algorithms(K-means,Hierarchical etc)
5. K-Nearest Neighbors (KNN)
6. SVM

#### Algorithms which do not require the use of feature scaling -
1. Decision Trees
2. Random Forest
3. Boosting Techniques(Gradient Boost, AdaBoost, XGBoost)
4. Naive Bayes

#### Feature Scaling Techniques
1. Standardization
2. Normalization

### Standardization

The formula used for standardization is :<br>
**z = (x - u) / s**  , where x- data point ,u-  mean, s - standard deviation

#### Normalization

Normalization, also sometimes referred to as min-max scaling, is a technique for scaling features in a dataset to a fixed range, between 0 and 1.


```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_normalized = scaler.fit_transform(df.drop(columns=["CHAS"]))
df_normalized=  pd.DataFrame(df_normalized,columns=df.drop(columns=["CHAS"]).columns)
```

```python
sns.histplot(x=df_normalized.LSTAT,kde=True)
```

When to use Min-Max Scaler vs Standardization?

- Though there is no rule of thumb on using normalization or standardization, *it has been observed that in most of the machine learning use-cases standardization performs well.* 
- *Normalization is often used in deep learning techniques.*
- You can try using both normalization and standardization and compare the performance of your model.
