# Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data strutrues.

```python
import seaborn as sns

%matplotlib
```

```python
help(sns.load_dataset)
```

```python
df = sns.load_dataset("tips")   # try using flights dataset
df
```

```python
sns.set_theme()
```

```python
# ba
```

```python
sns.set_theme(style = "whitegrid")
```

```python
sns.barplot(data = df , x = "sex",y = "tip")
```

```python
sns.barplot(data = df , x = "time",y = "total_bill")
```

```python
sns.barplot(data = df ,x = "time",y ="total_bill",hue= "sex") # try giving hue = "sex"
```

```python
#histogram
sns.histplot(data = df , x = "total_bill" )
```

```python
#histogram
sns.histplot(data = df , x = "total_bill" ,hue = "sex")
```

```python
# Scatter plot
sns.scatterplot(data = df , x = "total_bill",y = "tip",hue ="smoker") # try giving , hue = "sex"
```

```python
# Line Plot
sns.set_theme()
sns.lineplot(data = df , x = "total_bill",y = "tip" ,hue = "smoker")
```
