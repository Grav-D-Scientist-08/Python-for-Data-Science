# Baye's Theorem :-
- Before understanding the bayes theorem let us take a look at these concepts -
  - **Posterior Probability** :- The updated probability of events A after considering new evidence B. Its writter as P(A|b).
  - **Prior Probability** :- Your initial belief about the probability of an events A before considering any new evidence. Its written as P(A).
  - **Likelihood** :- The probability of observing the evidence B event A were true. Its written as P(B|A)
  - The bayes theorem states that 
     - P(A|B) = P(B|A) P(A)/P(B)
  - Now the question arises why to use this formula to calculate the conditional probability (Posterior probability) in stead of the traditional formula for conditional probability?
  - P(A|B) is the posterior probability of event A given the occurrence  of event B.
  - P(B|A) is the likelyhood or conditional porbability of event B given that event A has occurred.
  - P(A) is the prior probability of events A, Which is the probability of A before considering evidence B.
  - P(B) is the probability of event B, the marginal likehood of observing B.
